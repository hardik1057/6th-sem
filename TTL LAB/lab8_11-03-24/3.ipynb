```python
#Use seaborn to create to create a barplot for categories A,B and C with corresponding [3,7,5].
import seaborn as sns
import matplotlib.pyplot as plt

# Categories and their corresponding values
categories = ['A', 'B', 'C']
values = [3, 7, 5]

# Create a bar plot
sns.barplot(x=categories, y=values)

plt.xlabel('Category')
plt.ylabel('Value')
plt.title('Bar Plot')
plt.show()
```
