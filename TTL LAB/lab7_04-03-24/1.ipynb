```python
# an online learning platfrom wants to understand the distribution of the time spent by the students on competing a particular course module.
# design and create a histogram based on the time data to visualise the concentration of study durations and identify any patterns in the 
#learning behaviour 
import matplotlib.pyplot as plt
import numpy as np
study_durations = np.random.exponential(scale=30, size=200)  
# Create a histogram
plt.figure(figsize=(8, 6))
plt.hist(study_durations, bins=20, color="skyblue", edgecolor="black")
plt.xlabel("Study Duration (minutes)")
plt.ylabel("Frequency")
plt.title("Distribution of Study Durations")

mean_duration = np.mean(study_durations)
plt.axvline(mean_duration, color="red", linestyle="--", label=f"Mean Duration: {mean_duration:.2f} min")

plt.legend()
plt.grid(axis="y")
plt.show()
```
