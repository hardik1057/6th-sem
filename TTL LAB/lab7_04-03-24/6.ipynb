```python
#a weather station records daily temp over a month. create a line plot to visualize the temperature trends
import matplotlib.pyplot as plt
import numpy as np

daily_temps = np.random.normal(loc=20, scale=5, size=30)  
plt.figure(figsize=(10, 5))
plt.plot(range(1, 31), daily_temps)

plt.title('Daily Temperature Over a Month')
plt.xlabel('Day')
plt.ylabel('Temperature (C)')

plt.show()
```
