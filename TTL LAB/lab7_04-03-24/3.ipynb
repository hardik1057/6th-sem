```python
# retial store wants to analyse the sales performance of its products accross multiple categories .Generate a barplot to compare the total
# sales for each product category. highlight the top performance categories and provide insight to enhance the sotre product offer.
import matplotlib.pyplot as plt
categories = ["Electronics", "Clothing", "Home Goods", "Toys", "Books"]
sales = [120000, 9000, 75000, 60000, 45000]
plt.figure(figsize=(10, 6))
plt.bar(categories, sales, color="skyblue")
plt.xlabel("Product Category")
plt.ylabel("Total Sales")
plt.title("Sales Performance by Product Category")
top_categories = [categories[i] for i in range(len(sales)) if sales[i] >= max(sales) * 0.8]
for i in range(len(categories)):
    if categories[i] in top_categories:
        plt.text(i, sales[i] + 5000, f"{sales[i]:,.0f}", ha="center", va="bottom", fontweight="bold")
plt.annotate("Enhance product offer in Electronics and Clothing", xy=(0.5, 0.9), xycoords="axes fraction", fontsize=12, color="red")
plt.tight_layout()
plt.show()
```
