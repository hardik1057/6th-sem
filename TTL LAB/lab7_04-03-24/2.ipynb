```python
#a manufacturing company is concerned about the production times of two different machines. cmompare the distribution of the production times 
#for each machine using boxplots. identify any outlier and access the central tendency in the production process.
import matplotlib.pyplot as plt
import numpy as np
machine1_times = np.random.normal(loc=10, scale=2, size=100) 
machine2_times = np.random.normal(loc=12, scale=3, size=100) 
plt.figure(figsize=(8, 6))
plt.boxplot([machine1_times, machine2_times], labels=["Machine 1", "Machine 2"])
plt.xlabel("Machine")
plt.ylabel("Production Time")
plt.title("Distribution of Production Times for Each Machine")
def find_outliers(data):
    q1 = np.percentile(data, 25)
    q3 = np.percentile(data, 75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    outliers = [x for x in data if x < lower_bound or x > upper_bound]
    return outliers

outliers_machine1 = find_outliers(machine1_times)
outliers_machine2 = find_outliers(machine2_times)

print(f"Outliers in Machine 1 production times: {outliers_machine1}")
print(f"Outliers in Machine 2 production times: {outliers_machine2}")

mean_machine1 = np.mean(machine1_times)
mean_machine2 = np.mean(machine2_times)
print(f"Mean production time for Machine 1: {mean_machine1:.2f}")
print(f"Mean production time for Machine 2: {mean_machine2:.2f}")

plt.show()
```
