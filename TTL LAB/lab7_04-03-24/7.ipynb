```python
#analyze the income distribution accorss different job categories using a box plot. how does the plot provide detailled information about the 
# distribution tails and variation? Do analyse them 
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

df = pd.DataFrame({
    'Job Category': np.random.choice(['Job A', 'Job B', 'Job C', 'Job D', 'Job E'], 1000),  # random job categories
    'Income': np.random.normal(loc=50000, scale=10000, size=1000)  # random income data
})

# Create a box plot
plt.figure(figsize=(10, 8))
sns.boxplot(x='Job Category', y='Income', data=df)

# Add title and labels
plt.title('Income Distribution Across Job Categories')
plt.xlabel('Job Category')
plt.ylabel('Income')

# Show the plot
plt.show()
```
