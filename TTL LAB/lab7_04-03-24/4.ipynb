```python
#a car delearship is interested in understading the relationship between the mileage and reasale value of the used cars in inventory. 
# collect data on the mileage and resale the value of the various car models and create a scatterplot. analyse the plot to identify 
# any correlation patterns and provide insights into mileage influences resale value, helping the dealership make informed pricing decision.
import matplotlib.pyplot as plt
import numpy as np

# Generate random data
np.random.seed(0)  # for reproducibility
mileage = np.random.randint(5000, 200000, size=100)  # random mileage between 5000 and 200000
resale_value = np.random.randint(50000, 300000, size=100)  # random resale value between 50000 and 300000

# Create a scatter plot
plt.scatter(mileage, resale_value)

# Add title and labels
plt.title('Mileage vs Resale Value')
plt.xlabel('Mileage')
plt.ylabel('Resale Value')

# Calculate and print the correlation
correlation_matrix = np.corrcoef(mileage, resale_value)
correlation_xy = correlation_matrix[0,1]
r_squared = correlation_xy**2

print(f"The correlation between mileage and resale value is {correlation_xy}")
print(f"The R-squared value is {r_squared}")

# Show the plot
plt.show()

```
